```python
import numpy as np
import cv2
import matplotlib
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import skimage
import scipy
from scipy import ndimage
from scipy import signal
%matplotlib inline
```

### Function for Reading Images

```python
def read_image(IMG_NAME):
    img = mpimg.imread(IMG_NAME)
    return img
```

### Gray Scale Hybrid Images

```python
I1 = read_image('c2.jpg')
I1 = cv2.cvtColor(I1, cv2.COLOR_BGR2GRAY)/255

J1 = read_image('c1.jpg')
J1 = cv2.cvtColor(J1, cv2.COLOR_BGR2GRAY)/255

I2 = scipy.ndimage.gaussian_filter(I1,7)
J2 = J1 - scipy.ndimage.gaussian_filter(J1,5)

image = 0.5*I2 + 0.5*J2

fig1=plt.figure(figsize=[1.5,1.5],dpi=300)
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.title('Hybrid Image (Zoom In)', fontsize=4)
fig1.savefig('1-1.svg',format='svg')

fig2=plt.figure(figsize=[1,1],dpi=300)
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.title('Hybrid Image (Zoom Out)', fontsize=4)
fig2.savefig('1-2.svg',format='svg')

fig3=plt.figure(figsize=[1.5,1.5],dpi=300)
plt.imshow(I2, cmap='gray')
plt.axis('off')
plt.title('Filtered Image', fontsize=4)
fig3.savefig('a1-1.svg',format='svg')

fig4=plt.figure(figsize=[1.5,1.5],dpi=300)
plt.imshow(J2, cmap='gray')
plt.axis('off')
plt.title('Filtered Image', fontsize=4)
fig4.savefig('a1-2.svg',format='svg')
```
